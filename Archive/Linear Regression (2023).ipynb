```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
plt.style.use('Solarize_Light2')
plt.rcParams["figure.dpi"] = 100 
```

```python
import sklearn.datasets
```

```python
data = sklearn.datasets.fetch_california_housing()
```

```python
california_housing_df = pd.concat([pd.DataFrame(data['data'] ,columns=data['feature_names']), pd.DataFrame(data['target'] ,columns=data['target_names'])], axis=1)
```

```python
california_housing_df.columns = ["median income", "house age", "average rooms", "average bedrooms", "population", "average occupants", "latitude", "longitude", "house value"]
```

```python
normalized_df = pd.DataFrame(columns=california_housing_df.columns)
for column in california_housing_df.columns:
    maximum = california_housing_df[column].max()
    normalized_df[column] = california_housing_df[column] / maximum
```

```python
X = california_housing_df[normalized_df.columns[:-1]]
y = california_housing_df[normalized_df.columns[-1]]
```

```python
m = X.shape[0]
n = X.shape[1]
```

```python
cols = (n // 2) + (n % 2)
rows = cols // 2

fig, axes = plt.subplots(rows, cols, figsize=(10, 5))

title_font = font = {'family': 'monospace', 'fontweight': 'bold', 'fontsize': 10}
label_font = {'family': 'monospace', 'fontweight': 'bold', 'fontsize': 8}

for ax, col in zip(axes.flatten(), X.columns):
    ax.scatter(X[col], y)
    ax.set_xlabel(col, fontdict=label_font)
    ax.set_ylabel('price', fontdict=label_font)
    ax.set_title(f'price vs {col}', fontdict=title_font)
    
plt.tight_layout()
```

All equations used to wite the LinearRegression class: 
$$ f_{\mathbf{w},b}(\mathbf{x}) = \mathbf{w} \cdot \mathbf{x} + b  \tag{1} \newline$$ 
$$
\begin{align}
\frac{\partial J(\mathbf{w},b)}{\partial w_j}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)})x_{j}^{(i)} \tag{2} \\
\frac{\partial J(\mathbf{w},b)}{\partial b}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)}) \tag{3} \newline
\end{align}
$$
$$\begin{align*}
& w_j\ \ = w_j -  \alpha \frac{\partial J(\mathbf{w},b)}{\partial b}\tag{4}  \newline\newline
& b\ \ = b -  \alpha \frac{\partial J(\mathbf{w},b)}{\partial b} \tag{5} 
\end{align*}$$

```python
class LinearRegression:
    def __init__(self, X, y, alpha=0.0001, epochs=10_000):
        self.m = X.shape[0]
        self.n = X.shape[1]
        
        self.X = X.to_numpy().reshape(self.n, self.m)
        self.y = y.to_numpy().reshape(1, self.m)
        
        self.alpha = alpha
        self.epochs = epochs
        
        self.w = X.values.mean(axis=0)
        self.b = 0
    
    def predict(self, x):
        y_hat = np.dot(self.w.T, x) + self.b
        return y_hat

    def gradient(self):
        error = self.y - self.predict(self.X)
        dw = (-2/self.m) * np.sum(self.X * error)
        db = (-2/self.m) * np.sum((error))
        return dw, db
    
    def fit(self):
        for _ in range(self.epochs):
            dw, db = self.gradient()
            self.w -= self.alpha * dw
            self.b -= self.alpha * db
```

```python
model = LinearRegression(X, y)
model.fit()
```
