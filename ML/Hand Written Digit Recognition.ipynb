# Hand Written Digit Recognition Project

```python
# Standard imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## importing and processing the dataset 
- this is the MINST dataset
- with 70,000 images of hand written digits

```python
# Importing the MNIST dataset from TensorFlow's Keras datasets module
# The MNIST dataset is a large database of handwritten digits commonly used for training image processing systems
from tensorflow.keras.datasets import mnist
```

```python
# Loading MNIST dataset
train, test = mnist.load_data()

# Unpacking training and testing data (images and labels)
train_images, train_labels = train
test_images, test_labels = test
```

```python
# Flattening the 2D images into 1D arrays for the KNN algorithm
flattend_train_images = train_images.reshape((train_images.shape[0], -1))
flattend_test_images = test_images.reshape((test_images.shape[0], -1))
```

```python
# Concatenating training and testing datasets into one
all_labels = np.concatenate([train_labels, test_labels])
all_flattend_images = np.concatenate([flattend_train_images, flattend_test_images])
all_images = np.concatenate([train_images, test_images])
```

```python
# Creating a Pandas DataFrame for better analysis
# Each row in the DataFrame represents a digit with its corresponding label and flattened image
digits = pd.DataFrame(zip(all_labels, all_flattend_images), columns=['labels', 'images'])

# Displaying the first 5 rows of the DataFrame
digits.head()
```

## Part 2

### digit frequency histogram

```python
# Creating a histogram to show the frequency of each digit in the dataset
plt.hist(all_labels, bins=10, range=(0,10), rwidth=0.8)

# Setting the y-axis limits to focus on the data range
plt.ylim(6000, 8000)

# Setting the title and axis labels for the plot
plt.title('Digit Frequency Histogram')
plt.xlabel('Digits')
plt.ylabel('Frequency')

# Displaying the plot
plt.show()
```

### showing the mean image of each number
#### objective:
- Are there any noticeable patterns in the drawing styles of digits among individuals?

```python
# Creating a figure with 10 subplots (2 rows and 5 columns) to display the mean image of each digit
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(8, 3))

# Looping over each digit from 0 to 9
for digit in range(10):
    # Computing the mean image of the current digit
    mean_image = np.mean(all_images[all_labels == digit], axis=0)
    
    # Selecting the appropriate subplot for the current digit
    ax = axes[digit // 5][digit % 5]
    
    # Displaying the mean image and setting the title
    ax.imshow(mean_image, cmap='jet')
    ax.set_title(f'Number: {digit}')
    
    # Hiding the axes
    ax.axis('off')

# Adjusting the layout for better visualization
plt.tight_layout()

# Displaying the plot
plt.show()
```

### the distribution of the distance from the mean of each image
#### objective:
- Are the digits consistently drawn in a similar manner?
- Is there a significant variability in the way the digits are drawn?

```python
# Importing required modules
from scipy.spatial import distance
import matplotlib.pyplot as plt

# Creating a figure with 10 subplots (2 rows and 5 columns) to display distance histograms
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(15, 6))

# Looping over each digit from 0 to 9
for digit in range(10):
    # Extracting all images of the current digit and computing their mean
    digit_images = all_flattend_images[all_labels == digit]
    mean_image = np.mean(digit_images, axis=0)
    
    # Computing Euclidean distance between each image of the current digit and the mean image
    distances = distance.cdist(digit_images, mean_image.reshape(1, -1), metric='euclidean')
    
    # Selecting the appropriate subplot for the current digit and plotting a histogram of distances
    ax = axes[digit // 5][digit % 5]
    ax.hist(distances, bins=50)
    ax.set_title(f'Digit {digit}')  # Setting the title to the current digit

# Adjusting the layout for better visualization
plt.tight_layout()

# Displaying the plot
plt.show()
```

## Part 3

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
print(f'train data length: {len(train_labels)}')
print(f'test data length: {len(test_labels)}')
```

### finding the optimal scaler for the MINST dataset

```python
# Importing different types of scalers from sklearn.preprocessing
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler, Normalizer

# Defining a list of scalers and their corresponding names
scalers = [None, StandardScaler, MinMaxScaler, RobustScaler, Normalizer]
names = ['NoScaler', 'StandardScaler', 'MinMaxScaler', 'RobustScaler', 'Normalizer']

# Initializing a dictionary to store accuracy for each scaler
scaler_dict = {}

# Initializing a list to store performance results
scaler_performance = []

# Looping over each scaler
for Scaler, name in zip(scalers, names):
    if Scaler != None:  # If the scaler is not None
        # Instantiate the scaler
        scaler = Scaler()
        
        # Scale the training and testing images
        scaled_train_images = scaler.fit_transform(flattend_train_images)
        scaled_test_images = scaler.fit_transform(flattend_test_images)
    
        # Instantiate the KNN classifier and fit it to the scaled training data
        knn = KNeighborsClassifier(n_neighbors=3)
        knn.fit(scaled_train_images, train_labels)
        
        # Predict the labels of the scaled testing images
        predicted_labels = knn.predict(scaled_test_images)
        
    else:  # If no scaler is used
        # Instantiate the KNN classifier and fit it to the original training data
        knn = KNeighborsClassifier(n_neighbors=3)
        knn.fit(flattend_train_images, train_labels)
        
        # Predict the labels of the original testing images
        predicted_labels = knn.predict(flattend_test_images)
    
    # Compute the accuracy of the model
    accuracy = accuracy_score(test_labels, predicted_labels)
    
    # Store the accuracy in the dictionary
    scaler_dict[name] = accuracy
    
    # Append the accuracy to the performance list
    scaler_performance.append(accuracy)
```

```python
# Printing the results of the predictions
scaler_dict
```

```python
# Selecting the best scaler based on the previous performance results
# np.argmax(scaler_performance) returns the index of the highest accuracy in the list
# scalers[np.argmax(scaler_performance)]() instantiates the best scaler
scaler = scalers[np.argmax(scaler_performance)]()

# Using the best scaler to transform the training and testing images
scaled_train_images = scaler.fit_transform(flattend_train_images)
scaled_test_images = scaler.fit_transform(flattend_test_images)
```

### finding the optimal k for the MINST dataset

```python
# Initializing an empty list to store accuracy scores
accuracy_list = []

# Defining a range of potential numbers of neighbors (k values) to test
n_neighbors = range(1, 11)

# For each k value in the specified range
for k in n_neighbors:
    # Initialize a KNN classifier with the current k value
    knn = KNeighborsClassifier(n_neighbors=k)
    
    # Train the KNN classifier on the scaled training images
    knn.fit(scaled_train_images, train_labels)
    
    # Use the trained classifier to predict labels for the scaled test images
    predicted_labels = knn.predict(scaled_test_images)
    
    # Calculate the accuracy of the classifier's predictions
    accuracy = accuracy_score(test_labels, predicted_labels)
    
    # Append the accuracy score to the list of accuracy scores
    accuracy_list.append(accuracy)

# Creating a plot to visualize the relationship between k values and accuracy
plt.figure(figsize=(8, 4))
plt.plot(n_neighbors, accuracy_list, marker='o', linestyle='-')
plt.title('Accuracy of K-Nearest Neighbors for Different Values of k')
plt.xlabel('k')
plt.ylabel('Accuracy')

# Displaying the plot
plt.show()
```

### more important metrics

```python
# Importing necessary metrics from sklearn
from sklearn.metrics import classification_report, confusion_matrix

# Initializing the K-Nearest Neighbors (KNN) classifier with optimal value of k (n_neighbors=4)
knn = KNeighborsClassifier(n_neighbors=4)

# Training the KNN model on the scaled training images
knn.fit(scaled_train_images, train_labels)

# Using the trained model to predict the labels of the test images
predicted_labels = knn.predict(scaled_test_images)

# Generating a confusion matrix to evaluate the performance of the classifier
# The confusion matrix shows the number of correct and incorrect predictions made by the classifier, broken down by each class
cm = confusion_matrix(test_labels, predicted_labels)
```

```python
# plotting the confusion matrix
fig, ax = plt.subplots(figsize=(7, 5))
im = ax.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
ax.figure.colorbar(im, ax=ax)
ax.set(xticks=np.arange(cm.shape[1]),
       yticks=np.arange(cm.shape[0]),
       xticklabels=range(10), yticklabels=range(10),
       title='Confusion matrix for KNN with k=4',
       ylabel='True label',
       xlabel='Predicted label')

thresh = cm.max() / 2.
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        ax.text(j, i, format(cm[i, j], 'd'),
                ha="center", va="center",
                color="white" if cm[i, j] > thresh else "black")
fig.tight_layout()
plt.show()
```
