### 2.2 גרדיאנט דיסנט

בחלק זה, נתאים את פרמטרי הרגרסיה הליניארית $w$ למאגר הנתונים שלנו באמצעות גרדיאנט דיסנט.

נזכיר שההשערה $h_W(x)$ עבור רגרסיה ליניארית ניתנת על ידי המודל הקוי:

$$ h_W(x) = w_0 + w_1 x_1$$


---



<div class="alert alert-block alert-warning">

**הערת יישום:** כדי לכתוב ולטפל בהשערה כמכפלה פנימית של שני וקטורים, אנו מגדירים $x_0=1$ ואז מטפלים ב- $x = [x_0, x_1] = [1, x_1]$ כווקטור. הווקטור השני של המכפלה הפנימית יהיה $w = [w_0, w_1]$ עם זה אנו כותבים את ההשערה כדלקמן:

$$ h_W(x) = w_0 + w_1 x_1 = w^Tx$$

</div>


---



#### 2.2.1 משוואות עדכון

המטרה של רגרסיה ליניארית היא למזער את פונקציית העלות

$$ J(w) = \frac{1}{2m} \sum_{i=1}^m \left( h\_{w}(x^{(i)}) - y^{(i)}\right)^2$$

הפרמטרים של המודל לעיל הם ערכי $w_0, w_1$. אלו הם

הערכים שצריך להתאים כדי למזער את העלות $J(w)$. דרך אחת לעשות זאת היא באמצעות אלגוריתם גרדיאנט דיסנט.


$$ w_j = w_j - \alpha \frac{\partial J}{\partial w_j}      \text{    simultaneously update } w_j \text{ for all } j$$

הנגזרות החלקיות של $J$ לגבי כל פרמטר הן (כדי לראות איך לחשב נגזרות חלקיות אלה

[\`לחץ כאן\`](https://studio.youtube.com/video/XjPN1cP9ddk)):

$$\frac{\partial J}{\partial w_0} = \frac{1}{m} \sum_{i=1}^m \left( h_W(x^{(i)}) - y^{(i)}\right) \qquad$$

$$\frac{\partial J}{\partial w_1} = \frac{1}{m} \sum_{i=1}^m \left( h_W(x^{(i)}) - y^{(i)}\right)x_1^{(i)} \qquad$$

מכיוון שהגדרנו $x_0=1$, אנו יכולים לכתוב את הנגזרת החלקית הראשונה לעיל כך:

$$\frac{\partial J}{\partial w_0} = \frac{1}{m} \sum_{i=1}^m \left( h_W(x^{(i)}) - y^{(i)}\right)x_0^{(i)} \qquad$$

$$\frac{\partial J}{\partial w_1} = \frac{1}{m} \sum_{i=1}^m \left( h_W(x^{(i)}) - y^{(i)}\right)x_1^{(i)} \qquad$$

או לכתוב את שתי הנגזרות בצורה כללית:

$$\frac{\partial J}{\partial w_j} = \frac{1}{m} \sum_{i=1}^m \left( h_W(x^{(i)}) - y^{(i)}\right)x_j^{(i)} \qquad$$

עם זה ניתן לכתוב את משוואות צעד העדכון של אלגוריתם גרדיאנט דיסנט כדלקמן:

$$ \begin{equation}
\label{}
\boxed{w_j = w_j - \alpha \frac{1}{m} \sum_{i=1}^m \left( h_W(x^{(i)}) - y^{(i)}\right)x_j^{(i)} \qquad  \text{    simultaneously update } w_j \text{ for all } j}
\end{equation}$$

עם כל צעד של גרדיאנט דיסנט, הפרמטרים שלך $w_j$ מתקרבים יותר לערכים האופטימליים שישיגו את העלות הנמוכה ביותר $J(w)$.

</div>


````python
# Extract the X and y values from the 'pandas' dataframe to 'numpy' arrays
X = df.Population.values
y = df.Profit.values
````

#### 2.2.3 חישוב פונקציית העלות $J(w)$

 בחלק זה, ניישם את הפונקציה cost_function אשר מחשבת את העלות $J(w)$ כך שתוכל לבדוק את התכנסות יישום הגרדיאנט דיסנט שלך.
אנא השלם את הקוד הבא עבור הפונקציה cost_function אשר מחשבת את $J(w)$. בזמן שאתה עושה זאת, זכור שהמשתנים $X$ ו-$y$ אינם ערכי סקלר. $X$ הוא מטריצה ששורותיה מייצגות דוגמאות מהקבוצת האימון ו-$y$ הוא וקטור שכל אלמנט בו מייצג את הערך בשורה נתונה של $X$.

תזכורת:

$$ J(w) = \frac{1}{2m} \sum_{i=1}^m \left( h_{w}(x^{(i)}) - y^{(i)}\right)^2$$


<a id="computeCost"></a>
</div>

````python
def cost_function(X, y, W):
    """
    Compute cost for linear regression. Computes the cost of using W as the
    parameter for linear regression to fit the data points in X and y.

    Parameters
    ----------
    X : array_like
        The input dataset of shape (m x n+1), where m is the number of examples,
        and n is the number of features. We assume a vector of one's already
        appended to the features so we have n+1 columns.

    y : array_like
        The values of the function at each data point. This is a vector of
        shape (m, ).

    W : array_like
        The parameters for the regression function. This is a vector of
        shape (n+1, ).

    Returns
    -------
    J : float
        The value of the regression cost function.

    Instructions
    ------------
    Compute the cost of a particular choice of W.
    You should set J to the cost.
    """

    # initialize some useful values
    m = y.size  # number of training examples

    # You need to return the following variables correctly
    J = 0

    # ====================== YOUR CODE HERE =====================

    J = (1 / (2*m)) * sum((np.dot(X, W) - y) ** 2)

    # ===========================================================
    return J
````

<div dir="rtl">

הבא נבדוק את המימוש שלך של $ J(w)$ על $X$ ו-$y$ עם כמה $W$ מומצאים:

</div>

````python
J = cost_function(X, y, W=np.array([0, 0]))
print('With W = [0, 0] \nCost computed = %.2f' % J)
print('Expected cost value (approximately) 32.07\n')

# further testing of the cost function
J = cost_function(X, y, W=np.array([-1, 2]))
print('With W = [-1, 2]\nCost computed = %.2f' % J)
print('Expected cost value (approximately) 54.24')
````

#### 2.2.4 גרדיאנט דיסנט (Gradient Decent)


 בחלק זה, תיישם את הפונקציה gradient_descent. מבנה הלולאה כבר נכתב עבורך, ועליך רק לספק את העדכונים ל-$w$ בתוך כל איטרציה.

בזמן התכנות, ודא שאתה מבין למה אתה מנסה לבצע אופטימיזציה ומה מתעדכן. זכור שהפרמטר לעלות $J(w)$ הוא הוקטור $w$, ולא  $X$ ו-$y$. כלומר, אנו ממזערים את ערך $J(w)$ על ידי שינוי ערכי הוקטור $w$, ולא על ידי שינוי $X$ או $y$. התייחס למשוואה במסגרת למעלה אם אתה לא בטוח. דרך טובה לאמת שהגרדיאנט דיסנט עובד כראוי היא להסתכל על הערך של $J(w)$ ולבדוק שהוא פוחת עם כל צעד.

הקוד הראשוני עבור הפונקציה gradientDescent קורא ל-cost_function בכל איטרציה ושומר את העלות לרשימה של פייתון. בהנחה שיישמת נכון את הגרדיאנט היורד ואת הפונקציה cost_function, הערך שלך עבור $J(w)$ לעולם לא אמור לעלות, וצריך להתכנס לערך יציב עד סוף האלגוריתם.

</div>




<div>
<a id="gradientDescent"></a>

---

<div dir="rtl">

<div class="alert alert-box alert-warning">

**וקטורים ומטריצות ב-** `numpy` **- הערות מימוש חשובות**

וקטור ב-`numpy` הוא מערך חד-ממדי, לדוגמה `np.array([1, 2, 3])` הוא וקטור. מטריצה ב-`numpy` היא מערך דו-ממדי, לדוגמה `np.array([[1, 2, 3], [4, 5, 6]])`. עם זאת, הערך הבא נחשב עדיין מטריצה `np.array([[1, 2, 3]])` מכיוון שיש לו שני ממדים, גם אם יש לו צורה של 1x3 (שנראית כמו וקטור).

בהינתן הנ"ל, לפונקציה `np.dot` אותה נשתמש לכל הכפלה של מטריצה/וקטור יש את המאפיינים הבאים:

- היא תמיד מבצעת מכפלות פנימיות על וקטורים. אם `x=np.array([1, 2, 3])`, אז `np.dot(x, x)` הוא סקלר.

- עבור הכפלת מטריצה-וקטור, אם $X$ היא מטריצה $m\times n$ ו-$y$ הוא וקטור באורך $m$, אז הפעולה `np.dot(y, X)` רואה את $y$ כוקטור $1 \times m$. מצד שני, אם $y$ הוא וקטור באורך $n$, אז הפעולה `np.dot(X, y)` רואה את $y$ כוקטור $n \times 1$.

</div>

---


````python
def gradient_descent(X, y, W, alpha, num_iters):
    """
    Performs gradient descent to learn `W`. Updates W by taking `num_iters`
    gradient steps with learning rate `alpha`.

    Parameters
    ----------
    X : array_like
        The input dataset of shape (m x n+1).

    y : array_like
        Value at given features. A vector of shape (m, ).

    W : array_like
        Initial values for the linear regression parameters.
        A vector of shape (n+1, ).

    alpha : float
        The learning rate.

    num_iters : int
        The number of iterations for gradient descent.

    Returns
    -------
    W : array_like
        The learned linear regression parameters. A vector of shape (n+1, ).

    J_history : list
        A python list for the values of the cost function after each iteration.

    Instructions
    ------------
    Peform a single gradient step on the parameter vector W.

    While debugging, it can be useful to print out the values of
    the cost function (computeCost) and gradient here.
    """
    # Initialize some useful values
    m = y.shape[0]  # number of training examples

    # make a copy of W, to avoid changing the original array, since numpy arrays
    # are passed by reference to functions
    W = W.copy()

    J_history = [] # Use a python list to save cost in every iteration

    for i in range(num_iters):
        # ==================== YOUR CODE HERE =================================
        
        W = W - alpha * (np.dot(X.T, ((np.dot(X, W)) - y)) / m)
        
        # =====================================================================

        # save the cost J in every iteration
        J_history.append(cost_function(X, y, W))

    return W, J_history
````

<div dir="rtl">

הבא נבדוק את פונקציית ה-gradient_descent  אותה מימשת ונדפיס את ה-$w$ שחושבו.

אנו מאתחלים את פרמטרי ה-$w$ ל-0 ואת קצב הלמידה $\alpha$ ל-0.01.

הרץ את התא הבא כדי לבדוק את הקוד שלך.

</div>

````python
# initialize fitting parameters
W = np.zeros(2)

# some gradient descent settings
iterations = 1500
alpha = 0.01

W, J_history = gradient_descent(X ,y, W, alpha, iterations)

#Let's extract the values of the W we found and round them to two decimal points
w_0 = round(W[0], 2)
w_1 = round(W[1], 2)
print(f"W found by gradient descent: {w_0},{w_1}")
print('Expected W values (approximately): [-3.63, 1.17]')

print(f"Hence the hypothesis function would be: h(x) = {round(W[0], 2)} + {round(W[1], 2)}*x_1")
````

### ויזואליזציה תלת מימדית של משטח פונקציית העלות $J(W)$
